```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading Data

```python
data = pd.read_csv('C:\\Users\\axscat\\Documents\\GitHub\\DisasterPrediction\\Merged Dataset\\Dataset.csv')
print(data.head())
```

Analysis 1: Identifying the most frequent event type in each state

```python
variables = ['Cooling_Days', 'Heating_Days', 'minTemp', 'MaxTemp', 'AverageTemp', 'Precipitation']
means_per_state = data.groupby('state')[variables].mean().reset_index()

def plot_variable_states(df, variable, ylabel):
    plt.figure(figsize=(15, 10))
    sns.barplot(x=variable, y='state', data=df, palette='viridis')
    plt.title(f'{variable} per State')
    plt.xlabel(variable)
    plt.ylabel('State')
    plt.show()

# Plotting each variable against states
for variable in variables:
    plot_variable_states(means_per_state, variable, 'State')
```
